```python
import os
import pandas as pd
import numpy as np

from supervised.automl import AutoML
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.model_selection import GroupKFold, LeaveOneGroupOut
```

```python
# --- 袩褍褌懈 ---
BASE_PATH = '/kaggle/input/lr1automl'
DATA_FILENAME = 'all_data_target.csv'
REPORT_FOLDER_NAME = 'mljar_results_quick'
file_path = os.path.join(BASE_PATH, DATA_FILENAME)
results_path = os.path.join("/kaggle/working/", REPORT_FOLDER_NAME)

print(f"袟邪谐褉褍蟹泻邪 写邪薪薪褘褏 懈蟹: {file_path}")
df = pd.read_csv(file_path)
df.columns = [c.strip() for c in df.columns]

# --- 袛邪薪薪褘械 ---
TARGET_COLUMN = 'Target' if 'Target' in df.columns else 'target'
assert TARGET_COLUMN in df.columns, "袧械 薪邪泄写械薪邪 泻芯谢芯薪泻邪 Target/target"

NON_FEATURE_COLS = ['Patient', 'task', 'Aggregate', TARGET_COLUMN]
X = df.drop(columns=NON_FEATURE_COLS, errors='ignore')
y = df[TARGET_COLUMN]
groups = df['Patient'] if 'Patient' in df.columns else None

print("肖芯褉屑邪 X:", X.shape, "| 褍薪懈泻邪谢褜薪褘褏 蟹薪邪褔械薪懈泄 y:", y.nunique())

# --- 孝懈锌 蟹邪写邪褔懈 懈 屑械褌褉懈泻邪 ---
def detect_task(y: pd.Series) -> str:
    if y.dtype.kind in "ifu":
        if y.nunique() <= 2:  return "binary_classification"
        if y.nunique() <= 20: return "multiclass_classification"
        return "regression"
    else:
        return "binary_classification" if y.nunique() == 2 else "multiclass_classification"

ml_task = detect_task(y)
eval_metric = {"regression":"rmse",
               "binary_classification":"accuracy",
               "multiclass_classification":"accuracy"}[ml_task]
print("ml_task:", ml_task, "| eval_metric:", eval_metric)

# --- 肖芯谢写褘 斜械蟹 褍褌械褔泻懈 锌芯 锌邪褑懈械薪褌邪屑 (袨袘携袟袗孝袝袥鞋袧袨 锌械褉械写邪写懈屑 胁 fit 泻邪泻 cv) ---
folds = None
if groups is not None:
    n_groups = pd.Series(groups).nunique()
    if n_groups >= 2:
        # 泻邪泻 胁 锌褉懈屑械褉械: 屑芯卸薪芯 LeaveOneGroupOut, 薪芯 斜褘褋褌褉械械 GroupKFold
        n_splits = min(5, n_groups) if n_groups >= 5 else n_groups
        if n_groups >= 5:
            splitter = GroupKFold(n_splits=n_splits)
            folds = list(splitter.split(X, y, groups=groups))
            print(f"袠褋锌芯谢褜蟹褍械屑 GroupKFold 褋 {n_splits} 褎芯谢写邪屑懈 锌芯 Patient (n_groups={n_groups})")
        else:
            logo = LeaveOneGroupOut()
            folds = list(logo.split(X, y, groups=groups))
            print(f"袠褋锌芯谢褜蟹褍械屑 LeaveOneGroupOut 锌芯 Patient (n_groups={n_groups})")

# validation_strategy 写谢褟 custom-褉械卸懈屑邪
if folds is not None:
    validation_strategy = {"validation_type": "custom"}
else:
    # 蟹邪锌邪褋薪芯泄 胁邪褉懈邪薪褌 (械褋谢懈 薪械褌 Patient)
    validation_strategy = {"validation_type": "stratified_kfold",
                           "n_folds": 5, "shuffle": True, "random_seed": 13}
print("validation_strategy:", validation_strategy)

# --- 袙械褋邪 泻谢邪褋褋芯胁 (褌芯谢褜泻芯 写谢褟 泻谢邪褋褋懈褎懈泻邪褑懈懈) ---
sample_weight = compute_sample_weight(class_weight="balanced", y=y) if "classification" in ml_task else None

# --- AutoML ---
algorithms = ["LightGBM", "Xgboost", "Random Forest", "CatBoost", "Linear", "Neural Network"]
print("-"*30)
print(f"袟邪锌褍褋泻 AutoML (Perform), 谢懈屑懈褌 {60*2} 褋械泻")
automl = AutoML(
    algorithms=algorithms,
    mode="Perform",
    ml_task=ml_task,
    eval_metric=eval_metric,
    results_path=results_path,
    total_time_limit=60*2,
    explain_level=2,
    validation_strategy=validation_strategy,
    random_state=13
)

# 袙袗袞袧袨: 械褋谢懈 懈褋锌芯谢褜蟹褍械屑 custom-胁邪谢懈写邪褑懈褞, 锌械褉械写邪褢屑 cv=folds!
fit_kwargs = dict(sample_weight=sample_weight)
if folds is not None:
    fit_kwargs["cv"] = folds

automl.fit(X, y, **fit_kwargs)

# --- 袥懈写械褉斜芯褉写 懈 褋芯褏褉邪薪械薪懈械 锌褉械写褋泻邪蟹邪薪懈泄 ---
lb = automl.get_leaderboard()
print("袥懈写械褉斜芯褉写 (褌芯锌-10):")
display(lb.head(10))

if "classification" in ml_task:
    preds = automl.predict(X)
    df_out = df.copy()
    df_out["pred"] = preds
    try:
        proba = automl.predict_proba(X)
        # 邪泻泻褍褉邪褌薪芯 胁芯褋褋褌邪薪芯胁懈屑 懈屑械薪邪 泻谢邪褋褋芯胁, 械褋谢懈 写芯褋褌褍锌薪褘
        labels = automl._get_model_info().get("labels", list(range(proba.shape[1])))
        proba_df = pd.DataFrame(proba, columns=[f"proba_{c}" for c in labels])
        df_out = pd.concat([df_out, proba_df], axis=1)
    except Exception:
        pass
else:
    df_out = df.copy()
    df_out["prediction"] = automl.predict(X)

out_path = os.path.join("/kaggle/working", "predictions_mljar.csv")
df_out.to_csv(out_path, index=False)
print(f"小芯褏褉邪薪械薪芯: {out_path}")
```
